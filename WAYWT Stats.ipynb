# WAYWT Stats

Dataset is WAYWT comments from Jan 2014 to Sept 1 2017. This is by no means a shining example of idiomatic pandas usage.

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_csv('waywt_comments.csv')
```

```python
df.head()
```

```python
by_author = df.groupby('author')
```

```python
upvote_total_by_author = pd.DataFrame(by_author['upvotes'].sum())
```

```python
upvote_total_by_author.sort_values(by='upvotes', ascending=False).head(20)
```

```python
direct_replies = pd.DataFrame(by_author['direct reply count'].sum()).sort_values(by='direct reply count', ascending=False)
```

```python
direct_replies.head(20)
```

```python
number_of_posts = pd.DataFrame(by_author['posting utc time'].count()).sort_values('posting utc time', ascending=False)
```

```python
number_of_posts.head(20)
```
